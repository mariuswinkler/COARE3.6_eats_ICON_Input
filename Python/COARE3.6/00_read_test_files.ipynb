## Preamble

```python
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pylab as plt
```

## Data

```python
# Read the ASCII file into a pandas DataFrame
file_path = 'test_36_output_withnowavesinput_withwarmlayer.txt'
df = pd.read_csv(file_path, delim_whitespace=True)

# Convert the DataFrame to an xarray Dataset
ds = xr.Dataset.from_dataframe(df)
```

```python
Cd = ds.Cd
```

```python
U10 = ds.U10
```

## Plotting

```python
SIZE = 15
plt.rcParams["axes.labelsize"]  = SIZE
plt.rcParams["legend.fontsize"] = SIZE
plt.rcParams["xtick.labelsize"] = SIZE
plt.rcParams["ytick.labelsize"] = SIZE
plt.rcParams["font.size"]       = SIZE
plt.rcParams["xtick.direction"] = 'in'
plt.rcParams["ytick.direction"] = 'in'
```

```python
fig, ax = plt.subplots()

ax.scatter(U10, Cd*1000, marker='.', s=1, color='k')

ax.set_xlabel('U10')
ax.set_ylabel(r'Cd / 10$^{-3}$')
ax.set_title(f'COARE3.6\n U10 vs Cd')

ax.set_xlim(0,13)
ax.set_ylim(0,2)

ax.spines['left'].set_position(('outward', 20))
ax.spines['bottom'].set_position(('outward', 20))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
